```python
#Etape 1 : Modélisation du texte comme succession de lettres
M = np.zeros((27,27))
#clean_text = 'sabah77 nada'
for i in range (1,len(clean_text)-1):
    if (clean_text[i] == " "):
        if (clean_text[i+1] == " "):
            M[0][0] += 1
        else :
            ordr_j = ord(clean_text[i+1])
            if (ordr_j > 96 and ordr_j < 123):
                M[0][ordr_j - 97+1] += 1
    else :
        ordr_i = ord(clean_text[i])
        if (ordr_i > 96 and ordr_i < 123):
            if (clean_text[i + 1] == " " ):
                M[ordr_i - 96][0] += 1
            else :
                ordr_j = ord(clean_text[i+1])
                if (ordr_j > 96 and ordr_j < 123):
                    M[ordr_i - 97+1][ordr_j - 97+1] += 1    
print(M)   
    
```

```python
def indice(lettre):
    if (ord(lettre) >=ord('a') and ord(lettre)<=ord('z')) or lettre  ==" " :
        if(lettre==" "):
            return 0
        else:
            asci=ord(lettre)-ord('a')+1
    else :
        asci = -1
    return asci
```

```python
def incre(text):
    M=np.zeros((27,27))
    for i in range(0,len(text)-1):
        if indice(text[i]) != -1 and indice(text[i+1]) != -1:
            M[indice(text[i])][indice(text[i+1])]=M[indice(text[i])][indice(text[i+1])]+1
    return M   
incre(out)
```

```python
#Matrice des decomptes de toutes les transitions
compteur = np.zeros(27)
for i in range (27):
    for j in range (27) :
        compteur[i] += M[i][j] 
print(compteur)

#Definition de la matrice de probabilite
P = np.zeros((27,27))
for i in range (27):
    for j in range (27):
        if (compteur[i] != 0):
            P[i][j] = M[i][j]/compteur[i]
print(P)
```

```python
#Etape 2 : Tester la performence de votre modèle 
score = 0
tentative = 0
for i in range(1, len(clean_text)-1):
    x = ord(clean_text[i])
    y = ord(clean_text[i + 1])
    if (x > 96 and x < 123 and y > 96 and y < 123):
        score += P[x - 97][y - 97]
        tentative += 1
fraction = score/tentative
fraction
```
