```python
def indicelettre(lettre):
    if((ord(lettre) >96 and ord(lettre)<=122) or lettre==" " ):
        if(lettre==" "):
            return 0
        else:
            asci=ord(lettre)-96
            return asci
    else:
        return -1
```

```python
def incrementer(text):
    matrice=np.zeros((27,27))
    for i in range(0,len(text)-1):
        if(indicelettre(text[i])==-1):
            i=i+1
        else:
            if(indicelettre(text[i+1])==-1):
                i=i+1
            else:
                matrice[indicelettre(text[i])][indicelettre(text[i+1])]=matrice[indicelettre(text[i])][indicelettre(text[i+1])]+1
    return matrice      
```

```python
M=np.zeros((27,27))
M=incrementer(out)
print(M)
```

```python
#Matrice des decomptes de toutes les transitions
compteur = np.zeros(27)
for i in range (27):
    for j in range (27) :
        compteur[i] += M[i][j] 
print(compteur)

#Definition de la matrice de probabilite
P = np.zeros((27,27))
for i in range (27):
    for j in range (27):
        P[i][j] = M[i][j]/compteur[i]
print(P)
```

```python
#Etape 2 : Tester la performence de votre modèle 

def proba(text):
    score = 0
    tentative = 0
    for i in range(0,len(text)-1):
        if(indicelettre(text[i])==-1):
            i=i+1
        else:
            if(indicelettre(text[i+1])==-1):
                i=i+1
            else:
                score += P[indicelettre(text[i])][indicelettre(text[i+1])]
                tentative += 1
    return score, tentative
score, tentative = proba(text)
print(score, tentative)
fraction = score/tentative
fraction
```

```python
def mot(liste, mat):
    M = np.zeros((len(mat),len(mat)))
    for i in range (0, len(liste)-1):
        m1 = mat.index(liste[i])
        m2 = mat.index(liste[i+1])
        M[m1][m2] += 1
    return M       
```
