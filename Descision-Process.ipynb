```python
import numpy as np 

#~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~. 
#                      Initialisation 
#~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~.~. 

gamma = 0.9
green = (0,3)
red = (1,3)

S = { (i, j) for j in range(4) for i in range(3) } 
rewarads = { s:0 for s in S }
rewarads[green] = 1 # case verte
rewarads[red] = -100 # case rouge
actions = { s:('U', 'D', 'L', 'R') for s in S} 
##actions[ (0,0) ] = ('D', 'R')
##actions[ (2,3) ] = ('U', 'L')
politique = { s: np.random.choice( actions[s] ) for s in S}
V = { s:0 for s in S} 
print(actions)
V[green] = 1
V[red] = -100
```

```python
actions[ (0,0) ] = ('D', 'R')
actions[ (0,1) ] = ('L', 'R')
actions[ (0,2) ] = ('D', 'L', 'R')
actions[ (0,3) ] = ('L', 'D')

actions[(1,0)] = ('U', 'D')
actions[(1,1)] = ( )
actions[(1,2)] = ('U', 'D','R')
actions[(1,3)] = ('U', 'D','L')

actions[(2,0)] = ('U', 'R')
actions[(2,1)] = ('L', 'R')
actions[(2,2)] = ('U', 'L', 'R')
actions[(2,3)] = ('L', 'U')
```

```python
#La fonction transition permet de passer d'une position actuelle a une autre grace a une action donnee
def transition(s, action):
    pos = s[0]
    pos_ = s[1]
    if (action == 'U'):
        if (pos == 2 and pos_ == 1):
            return s
        if (pos == 0):
            pos = 0
        else :
            pos -= 1
    if (action == 'D'):
        if (pos == 0 and pos_ == 1):
            return s
        if (pos == 2):
            pos = 2
        else :
            pos += 1
    if (action == 'L'):
        if (pos == 1 and pos_ == 2):
            return s
        if (pos_ == 0):
            pos_ = 0
        else :
            pos_ -= 1
    if (action == 'R'):
        if (pos == 1 and pos_ == 0):
            return s
        if (pos_ == 3):
            pos_ = 3
        else :
            pos_ += 1
    return (pos, pos_)


#permet de donner la position de la case perpendiculaire a droite
def direction_perpend_d(pos, ligne, col):
    if (pos == 'U' or pos == 'D'):
        if (ligne == 1 and col == 0):
            return (ligne, col)
        if col == 3:
            return (ligne, 3)
        else :
            return (ligne, col + 1)
        
    if (pos == 'L' or pos == 'R'):
        if (ligne == 2 and col == 1):
            return (ligne, col)
        if ligne == 0:
            return (0, col)
        else :
            return (ligne - 1, col)

#permet de donner la position de la case perpendiculaire a gauche
def direction_perpend_g(pos, ligne, col):
    if (pos == 'U' or pos == 'D'):
        if (ligne == 1 and col == 2):
            return (ligne, col)
        if col == 0:
            return (ligne, 0)
        else :
            return (ligne, col - 1)
        
    if (pos == 'L' or pos == 'R'):
        if (ligne == 0 and col == 1):
            return (ligne, col)
        if ligne == 2:
            return (2, col)
        else :
            return (ligne + 1, col)
```

```python
#la fonction permet le calcul des cases avec les valeurs optimal a chaque iteration
def calcul_etat(iter):
    d = ['U', 'D', 'L', 'R']
    W = V.copy()
    Z = V.copy()
    politique = []
    for itr in range (iter):
        for i in W : 
            Vopt = []
            for p in d:
                t = transition(i,p)
                Vopt.append(0.8*Z[t] + 0.1*Z[direction_perpend_d(p, i[0], i[1])] + 0.1*Z[direction_perpend_g(p, i[0], i[1])] ) 
            W[i] =  rewarads[i] + gamma * max(Vopt)
            ind = Vopt.index(max(Vopt))
            politique.append(d[ind])
            if i == (1,1):
                W[(1,1)] = 0
        Z = W.copy()
    print("La politique suivie :",politique,"\n")
    return W

itr = 5
etat = calcul_etat(itr)
print ("Pour un nombre d'iterations = ",itr," : \n")

for i in etat :
    print(etat[i]," ",i,"\n")
    
```
